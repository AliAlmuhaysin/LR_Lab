## EDA (Exploratory Data Analysis)

## Q1: Create a dataframe and Save that dataset inside it.

```python
BostonData = pd.DataFrame(boston.data, columns=boston.feature_names)
BostonData['Price'] = boston.target
```

## Q2: Print the head rows of the dataframe.

```python
BostonData.head()
```

## Q3: Use histogram to show the distribution of House Prices.

```python
sns.histplot(data=BostonData, x='Price')
plt.show()
```

## Q4: Use a heatmap to show the correlation between features and the target labels.

```python
plt.figure(figsize=(12,10), dpi=96)
sns.heatmap(BostonData.corr(), annot=True, vmin=-1, vmax=1)
plt.show()
```

## Q5: Use a lmplot to draw the relations between price and LSTAT.

```python
sns.lmplot(data = BostonData, x='Price', y='LSTAT', height=6, aspect=2/1)
plt.show()
```

## Q6: Use a lmplot to draw the relations between price and RM.

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
X = BostonData[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = BostonData['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
LinModel = LinearRegression()
```

## Q9: Train the Model.

```python
LinModel.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(LinModel.intercept_)
```

```python
CoeffDf = pd.DataFrame(LinModel.coef_,X.columns,columns=['Coefficient'])
CoeffDf
```

```python
Predict = LinModel.predict(X_test)
Predict
```

```python
RealVal = np.array(y_test)
RealVal
```

```python
sns.scatterplot(x=Predict, y=RealVal)
```

## Residual Histogram

```python
sns.distplot((y_test-Predict));
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, Predict))
print('MSE:', metrics.mean_squared_error(y_test, Predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, Predict)))
```

```python
print('R^2:', metrics.r2_score(y_test, Predict))
```
